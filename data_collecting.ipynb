```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
ilçe = input("İlan bilgilerini almak istediğiniz ilçeyi yazınız: ")
trToEng = str.maketrans("ÇĞİÖŞÜçğıöşü", "CGIOSUcgiosu")
ilçe = ilçe.translate(trToEng).lower()

print("BU İŞLEM BİRAZ UZUN SÜREBİLİR!")
```

```python
def grab_advert(city):
    
    driver = webdriver.Firefox()
    url = "https://www.sahibinden.com/satilik/istanbul"
    driver.maximize_window()
    driver.get(url + '-' + city)
    time.sleep(5)
    
    
    KAPAT = driver.find_element_by_css_selector(".onetrust-close-btn-handler")
    KAPAT.click()
    KAPAT = driver.find_element_by_css_selector(".feature-discovery__icon-circle")
    KAPAT.click()
    time.sleep(1)

    
    ilanNo = 1
    sayfaSayisi = 1
    daire_ilanBilgileri = []
    residence_ilanBilgileri = []
    müstakilEv_ilanBilgileri = []
    villa_ilanBilgileri = []

    while (ilanNo<23):
        
        try:
            
            if(driver.find_element_by_xpath('//html//body//div[5]//div[4]//form//div//div[3]//table//tbody//tr[' + str(ilanNo) + ']//td[2]').text not in ["Daire","Residence","Müstakil Ev","Villa"]):
                ilanNo += 1
                
            else: 
                time.sleep(1)
                konutTipi = driver.find_element_by_xpath('//html//body//div[5]//div[4]//form//div//div[3]//table//tbody//tr[' + str(ilanNo) + ']//td[2]').text
                time.sleep(1)
                driver.find_element_by_xpath('//html//body//div[5]//div[4]//form//div//div[3]//table//tbody//tr[' + str(ilanNo) + ']//td[2]').click()
                time.sleep(1)
                
                if(konutTipi == "Daire"):
                    print("{}. sayfadaki {}. ilandayım".format(sayfaSayisi, ilanNo))
                    
                    time.sleep(1)
                    fiyat = driver.find_element_by_xpath('/html/body/div[5]/div[4]/div[1]/div/div[2]/div[2]/h3').text
                    time.sleep(1)
                    mahalle = driver.find_element_by_xpath('//div[@class="classifiedInfo "]//h2//a[3]').text
                    time.sleep(0.5)
                    detay = driver.find_elements_by_xpath('//div[@class="classifiedInfo "]//ul[@class="classifiedInfoList"]//li//span[@class=""]')

                    
                    ilanBilgisi = []
                    ilanBilgisi.append(konutTipi)
                    ilanBilgisi.append(mahalle)
                    for bilgi in detay:
                        ilanBilgisi.append(bilgi.text)
                    ilanBilgisi.append(fiyat)
                    daire_ilanBilgileri.append(ilanBilgisi)
                    del ilanBilgisi

                elif(konutTipi == "Residence"):
                    print("{}. sayfadaki {}. ilandayım".format(sayfaSayisi, ilanNo))
                    
                    time.sleep(1)
                    fiyat = driver.find_element_by_xpath('/html/body/div[5]/div[4]/div[1]/div/div[2]/div[2]/h3').text
                    time.sleep(1)
                    mahalle = driver.find_element_by_xpath('//div[@class="classifiedInfo "]//h2//a[3]').text
                    time.sleep(0.5)
                    detay = driver.find_elements_by_xpath('//div[@class="classifiedInfo "]//ul[@class="classifiedInfoList"]//li//span[@class=""]')

                    
                    ilanBilgisi = []
                    ilanBilgisi.append(konutTipi)
                    ilanBilgisi.append(mahalle)
                    for bilgi in detay:
                        ilanBilgisi.append(bilgi.text)
                    ilanBilgisi.append(fiyat)
                    residence_ilanBilgileri.append(ilanBilgisi)
                    del ilanBilgisi
  
                elif(konutTipi == "Müstakil Ev"):
                    print("{}. sayfadaki {}. ilandayım".format(sayfaSayisi, ilanNo))
                    
                    time.sleep(1)
                    fiyat = driver.find_element_by_xpath('/html/body/div[5]/div[4]/div[1]/div/div[2]/div[2]/h3').text
                    time.sleep(1)
                    mahalle = driver.find_element_by_xpath('//div[@class="classifiedInfo "]//h2//a[3]').text
                    time.sleep(0.5)
                    detay = driver.find_elements_by_xpath('//div[@class="classifiedInfo "]//ul[@class="classifiedInfoList"]//li//span[@class=""]')

                    
                    ilanBilgisi = []
                    ilanBilgisi.append(konutTipi)
                    ilanBilgisi.append(mahalle)
                    for bilgi in detay:
                        ilanBilgisi.append(bilgi.text)
                    ilanBilgisi.append(fiyat)
                    müstakilEv_ilanBilgileri.append(ilanBilgisi)
                    del ilanBilgisi
                    
                else:
                    print("{}. sayfadaki {}. ilandayım".format(sayfaSayisi, ilanNo))
                    
                    time.sleep(1)
                    fiyat = driver.find_element_by_xpath('/html/body/div[5]/div[4]/div[1]/div/div[2]/div[2]/h3').text
                    time.sleep(1)
                    mahalle = driver.find_element_by_xpath('//div[@class="classifiedInfo "]//h2//a[3]').text
                    time.sleep(0.5)
                    detay = driver.find_elements_by_xpath('//div[@class="classifiedInfo "]//ul[@class="classifiedInfoList"]//li//span[@class=""]')
                                            
                    ilanBilgisi = []
                    ilanBilgisi.append(konutTipi)
                    ilanBilgisi.append(mahalle)
                    for bilgi in detay:
                        ilanBilgisi.append(bilgi.text)
                    ilanBilgisi.append(fiyat)
                    villa_ilanBilgileri.append(ilanBilgisi)
                    del ilanBilgisi
            
            
                driver.back()
                time.sleep(0.5)
                driver.execute_script("window.scrollBy(0, 120)")
                ilanNo += 1
                      
        except:
            print("!!!!!!!!!!!!!!!!!!!!!!!!! İLAN AÇILAMADI !!!!!!!!!!!!!!!!!!!!!!!!!")
            ilanNo += 1
            
        if(ilanNo == 23 and sayfaSayisi<50):
            ilanNo = 1
            driver.get(url + '-' + city + '?pagingOffset=' + str(sayfaSayisi * 20))
            sayfaSayisi += 1
            time.sleep(1)
        else:
            continue

            
    driver.close()

    
    return daire_ilanBilgileri, residence_ilanBilgileri, müstakilEv_ilanBilgileri, villa_ilanBilgileri

daire_ilanBilgileri, residence_ilanBilgileri, müstakilEv_ilanBilgileri, villa_ilanBilgileri = grab_advert(ilçe)
```

```python
daire_columnsName = ["HOUSE_TYPE","NEIGHBORHOOD","GROSS_AREA","NET_AREA","NUMBER_OF_ROOMS","AGE_OF_BUİLDİNG",
                     "FLOOR_LOCATION","TOTALNUMBER_OF_FLOORS","TYPE_OF_HEATING","NUMBER_OF_BATHROOMS","IS_THERE_BALCONY",
                     "IS_IT_FURNISHED","WHO_IS_AT_HOME","ON_BUILDINGS","NAME_OF_BUILDINGS","DUES","AVAILABLE_FOR_CREDIT",
                     "TITLE_DEED","WHO_IS_THE_SELLER","IS_IT_VIEWABLE","PRICE"]
df_daire = pd.DataFrame(daire_ilanBilgileri, columns = daire_columnsName)



residence_columnsName = ["HOUSE_TYPE","NEIGHBORHOOD","GROSS_AREA","NET_AREA","NUMBER_OF_ROOMS","AGE_OF_BUİLDİNG",
                         "FLOOR_LOCATION","TOTALNUMBER_OF_FLOORS","TYPE_OF_HEATING","NUMBER_OF_BATHROOMS","IS_THERE_BALCONY",
                         "IS_IT_FURNISHED","WHO_IS_AT_HOME","ON_BUILDINGS","NAME_OF_BUILDINGS","DUES","AVAILABLE_FOR_CREDIT",
                         "TITLE_DEED","WHO_IS_THE_SELLER","IS_IT_VIEWABLE","PRICE"]
df_residence = pd.DataFrame(residence_ilanBilgileri,columns = residence_columnsName)



df_müstakilEv_columnsName = ["HOUSE_TYPE","NEIGHBORHOOD","GROSS_AREA","NET_AREA","NUMBER_OF_ROOMS","AGE_OF_BUİLDİNG",
                             "TOTALNUMBER_OF_FLOORS","TYPE_OF_HEATING","NUMBER_OF_BATHROOMS",
                             "IS_THERE_BALCONY","IS_IT_FURNISHED","WHO_IS_AT_HOME","ON_BUILDINGS","NAME_OF_BUILDINGS",
                             "DUES","AVAILABLE_FOR_CREDIT","TITLE_DEED","WHO_IS_THE_SELLER","IS_IT_VIEWABLE","PRICE"]
df_müstakilEv = pd.DataFrame(müstakilEv_ilanBilgileri,columns = df_müstakilEv_columnsName)



villa_columnsName = ["HOUSE_TYPE","NEIGHBORHOOD","GROSS_AREA","NET_AREA","NUMBER_OF_ROOMS","AGE_OF_BUİLDİNG",
                     "TOTALNUMBER_OF_FLOORS","TYPE_OF_HEATING","NUMBER_OF_BATHROOMS","IS_IT_FURNISHED","WHO_IS_AT_HOME",
                     "ON_BUILDINGS","NAME_OF_BUILDINGS","DUES","AVAILABLE_FOR_CREDIT","TITLE_DEED","WHO_IS_THE_SELLER",
                     "IS_IT_VIEWABLE","PRICE"]
df_villa = pd.DataFrame(villa_ilanBilgileri,columns = villa_columnsName)
```
